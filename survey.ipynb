```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
responses_df = pd.read_csv("responses.csv")
responses_df = responses_df.drop_duplicates(["email"], keep='last')
```

```python
figs_dir = "images"
```

# Number of unique participants

```python
responses_df.shape[0]
```

# Age groups

```python
age_groups = ["Under 20 years old", "20 - 29 years old", "30 - 39 years old", "40 - 49 years old", "50 - 59 years old", "60 years or older"]
y1 = [sum(responses_df["age"] == i) for i in age_groups]
y2 = [(val/responses_df.shape[0])*100 for val in y1]

fig, ax1 = plt.subplots()
plt.xticks(rotation=-20, ha="left")

ax1.set_xlabel('Age')
ax1.set_ylabel('Participants (%)')
ax1.bar(age_groups, y2)
ax1.tick_params(axis='y')

ax2 = ax1.twinx()  

ax2.set_ylabel('Number of participants')  
ax2.bar(age_groups, y1)
ax2.tick_params(axis='y')

plt.tight_layout()
# plt.savefig("Age.png", dpi=300)
```

# Gender groups

```python
gender_groups = ["Female", "Male", "Other", "Prefer not to answer."]
y1 = [sum(responses_df["gender"] == i) for i in gender_groups]
y2 = [(val/responses_df.shape[0])*100 for val in y1]

fig, ax1 = plt.subplots()

ax1.set_xlabel('Gender')
ax1.set_ylabel('Participants (%)')
ax1.bar(gender_groups, y2)
ax1.tick_params(axis='y')

ax2 = ax1.twinx()  

ax2.set_ylabel('Number of participants') 
ax2.bar(gender_groups, y1)
ax2.tick_params(axis='y')

plt.tight_layout()
# plt.savefig("Gender.png", dpi=300)
```

# Education


```python
education_groups = ["Less than a high school diploma", "High school diploma", "Bachelor's degree", "Master's degree", "Doctorate"]
y1 = [sum(responses_df["education"] == i) for i in education_groups]
y2 = [(val/responses_df.shape[0])*100 for val in y1]

fig, ax1 = plt.subplots()
plt.xticks(rotation=-20, ha="left")

ax1.set_xlabel('Education')
ax1.set_ylabel('Participants (%)')
ax1.bar(education_groups, y2)
ax1.tick_params(axis='y')

ax2 = ax1.twinx() 

ax2.set_ylabel('Number of participants')  
ax2.bar(education_groups, y1)
ax2.tick_params(axis='y')

plt.tight_layout()
# plt.savefig("Education.png", dpi=300)
```

# Main personal computer usage time

```python
pc1_time_df = responses_df[responses_df["pc1_usage_time"].notna()]
```

```python
pc1_time_df["pc1_usage_time"].mean()
```

```python
day_time_groups = range(25)
y1 = [sum(pc1_time_df["pc1_usage_time"] == i) for i in day_time_groups]
y2 = [(val/pc1_time_df.shape[0])*100 for val in y1]

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Number of hours main personal computer is used in a day')
ax1.set_ylabel('Participants (%)')
ax1.bar(day_time_groups, y2)
ax1.tick_params(axis='y')

ax2 = ax1.twinx()  

color = 'tab:blue'
ax2.set_ylabel('Number of participants') 
ax2.bar(day_time_groups, y1)
ax2.tick_params(axis='y')

plt.tight_layout()
# plt.savefig("Main_PC_Time.png", dpi=300)
```

# Main personal computer CPU utilization

```python
pc1_cpu_df = responses_df[responses_df["pc1_cpu"].notna()]
```

```python
pc1_cpu_df["pc1_cpu"].mean()
```

```python
cpu_groups = range(0, 100, 10)
x_ticks = ["0", "<=10", ">10 and <=20", ">20 and <=30", ">30 and <=40", ">40 and <=50", ">50 and <=60", 
           ">60 and <=70", ">70 and <=80", ">80 and <=90", ">90 and <=100"]
y1 = [sum((pc1_cpu_df['pc1_cpu'] == 0))]
for i in cpu_groups:
    y1.append(sum((pc1_cpu_df['pc1_cpu'] > i) & (pc1_cpu_df['pc1_cpu'] <= i+10)))
y2 = [(val/pc1_cpu_df.shape[0])*100 for val in y1]

fig, ax1 = plt.subplots()
plt.xticks(rotation=-30, ha="left")

ax1.set_xlabel('Main computer CPU utilization')
ax1.set_ylabel('Participants (%)')
ax1.bar(x_ticks, y2)
ax1.tick_params(axis='y')

ax2 = ax1.twinx() 

ax2.set_ylabel('Number of participants')  
ax2.bar(x_ticks, y1)
ax2.tick_params(axis='y')
```
